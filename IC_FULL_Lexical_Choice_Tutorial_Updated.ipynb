#### COCA (Corpus of Contemporary American English)
- A large, balanced corpus of American English.
- Useful for understanding word frequency and collocations.

**Use Case:**
COCA can be used to analyze how frequently certain words are used in contemporary American English, as well as to identify common collocations (words that frequently appear together). This can help language learners understand the most common and natural ways of expressing ideas in American English.

#### BNC (British National Corpus)
- A large corpus of British English.
- Provides insights into British English usage and frequency.

**Use Case:**
BNC is useful for studying the differences between British and American English. It can provide insights into word frequency, regional variations, and changes in language use over time. Language learners can use BNC to understand the nuances of British English and to compare it with other varieties of English.

#### LIWC (Linguistic Inquiry and Word Count)
- A text analysis tool that calculates the degree to which different categories of words are used in a text.
- Useful for psychological and social analysis of text.

**Use Case:**
LIWC can be used to analyze the emotional and psychological content of texts. It categorizes words into various psychological and social dimensions, such as emotion, cognition, and social relationships. Researchers can use LIWC to study the language patterns associated with different psychological states or social contexts.

```python
from gensim.models import Word2Vec


sentences = [
    ["I", "love", "natural", "language", "processing"],
    ["Word2Vec", "is", "a", "great", "tool"],
    ["This", "is", "an", "unsuperficial", "analysis", "of", "natural", "language"]
]

model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)


vector_natural = model.wv['natural']
vector_unsuperficial = model.wv['unsuperficial']

from numpy import dot
from numpy.linalg import norm

cosine_similarity = dot(vector_natural, vector_unsuperficial) / (norm(vector_natural) * norm(vector_unsuperficial))
print(f"Cosine similarity between 'natural' and 'unsuperficial': {cosine_similarity}")


similar_words_unsuperficial = model.wv.most_similar('unsuperficial')
print("\nWords similar to 'unsuperficial':")
for word, similarity in similar_words_unsuperficial:
    print(f"{word}: {similarity}")
```

```python
import nltk
from nltk.corpus import wordnet as wn
from gensim.models import Word2Vec
import numpy as np

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')


sentences = [
    ["I", "love", "natural", "language", "processing"],
    ["Word2Vec", "is", "a", "great", "tool"],
    ["This", "is", "an", "unsuperficial", "analysis", "of", "raw", "language"]
]


model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)


def get_wordnet_synonyms(word):
    synonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return synonyms


def enrich_embedding_with_wordnet(word, model):
    synonyms = get_wordnet_synonyms(word)
    print(f"Synonyms for '{word}': {synonyms}")
    if not synonyms:
        print(f"No synonyms found for '{word}' in WordNet.")
        return model.wv[word]

    synonym_vectors = [model.wv[syn] for syn in synonyms if syn in model.wv]
    print(f"Synonym vectors found for '{word}': {len(synonym_vectors)}")
    if not synonym_vectors:
        print(f"No synonym vectors found in the model for '{word}' synonyms.")
        return model.wv[word]

    enriched_vector = np.mean([model.wv[word]] + synonym_vectors, axis=0)
    return enriched_vector


def update_model_with_enriched_vector(word, enriched_vector, model):
    if word in model.wv:
        model.wv[word] = enriched_vector
    else:
        print(f"Word '{word}' not found in the model vocabulary.")


word = 'natural'
if word in model.wv:
    
    similar_words_before = model.wv.most_similar(word)
    print(f"Words similar to '{word}' before enrichment:")
    for similar_word, similarity in similar_words_before:
        print(f"{similar_word}: {similarity}")

    
    enriched_vector = enrich_embedding_with_wordnet(word, model)

    
    update_model_with_enriched_vector(word, enriched_vector, model)

    
    similar_words_after = model.wv.most_similar(word)
    print(f"\nWords similar to '{word}' after enrichment:")
    for similar_word, similarity in similar_words_after:
        print(f"{similar_word}: {similarity}")
else:
    print(f"Word '{word}' not found in the model vocabulary.")
```


#### GloVe
Generates word embeddings based on word co-occurrence statistics. Captures both local and global word relationships.


```python

import numpy as np

# Example using pre-trained GloVe embeddings
def load_glove_embeddings(file_path):
    embeddings_index = {}
    with open(file_path, encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    return embeddings_index

# Load GloVe embeddings (example file path)
glove_embeddings = load_glove_embeddings('glove.6B.200d.txt')
print(glove_embeddings['natural'])
```

```python
import pandas as pd
from IPython.display import display


glove_file_path = 'glove.6B.200d.txt'

def load_glove_embeddings(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()


    data = []
    for line in lines:
        values = line.split()
        word = values[0]
        vector = list(map(float, values[1:]))
        data.append([word] + vector)


    columns = ['word'] + [f'vector_{i}' for i in range(1, len(data[0]))]
    df = pd.DataFrame(data, columns=columns)
    return df


glove_df = load_glove_embeddings(glove_file_path)


print("Snapshot of GloVe Embeddings:")
display(glove_df.head())  
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


use_embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

sentences = [
    "I love natural language processing",
    "Word embeddings are very useful for NLP tasks"
]

use_embeddings = use_embed(sentences)
use_similarity = cosine_similarity(use_embeddings)

tokenized_sentences = [sentence.lower().split() for sentence in sentences]
word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=50, window=5, min_count=1, workers=4)

def average_word_vectors(sentence, model):
    words = sentence.lower().split()
    word_vectors = [model.wv[word] for word in words if word in model.wv]
    if not word_vectors:
        return np.zeros(model.vector_size)
    return np.mean(word_vectors, axis=0)

word2vec_embeddings = np.array([average_word_vectors(sentence, word2vec_model) for sentence in sentences])
word2vec_similarity = cosine_similarity(word2vec_embeddings)

print("Universal Sentence Encoder (USE) Embeddings:")
print(use_embeddings)
print("USE Cosine Similarity:")
print(use_similarity)

print("\nWord2Vec Embeddings (Averaged):")
print(word2vec_embeddings)
print("Word2Vec Cosine Similarity:")
print(word2vec_similarity)
```


#### BERT
Generates contextual embeddings for sentences.


```python

from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

sentence = "I love natural language processing"
inputs = tokenizer(sentence, return_tensors='pt')
outputs = model(**inputs)

embeddings = outputs.last_hidden_state[:, 0, :]
print(embeddings)
```

```python
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity


tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')


sentences = [
    "I am extremely happy",
    "Everybody is not sad today"
]


inputs = tokenizer(sentences, return_tensors='pt', padding=True, truncation=True)


with torch.no_grad():
    outputs = model(**inputs)


cls_embeddings = outputs.last_hidden_state[:, 0, :].numpy()


similarity = cosine_similarity(cls_embeddings)


print("BERT Embeddings for [CLS] token:")
print(cls_embeddings)
print("Cosine Similarity between sentences:")
print(similarity)
```


#### GPT-2
A large-scale generative model capable of understanding and generating human-like text.


```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, GPT2Config
import torch

model_name = 'gpt2'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

input_text = "I feel great today. Yesterday was not so good. The other month too. In fact, "


input_ids = tokenizer.encode(input_text, return_tensors='pt')

with torch.no_grad():
    outputs = model.generate(
        input_ids,
        max_length=100,   
        num_return_sequences=1,   
        pad_token_id=tokenizer.eos_token_id,   
        do_sample=True,   
        temperature=0.7,  
    )

for output in outputs:
    generated_text = tokenizer.decode(output, skip_special_tokens=True)
    print(generated_text)
```

## So What Scenarios of Lexical Choice Characteristics

### Synonym Variance in Discourse

**Scenario**: In a discourse, multiple synonyms are used interchangeably.

**Example**: A passage alternates between HAPPY,JOYFUL,CONTENT, and PLEASED.

**Category**: **Lexical Diversity**

**Characteristics**:
- **Synonym Usage**: High variance of synonymous words indicates a rich vocabulary and can make the text more engaging.
- **Implication**: Shows the effort of the writer or speaker to avoid repetition and keep the discourse lively.
- **Tool Insights**:
  - **WordNet**: Can identify synonyms and measure their usage.
  - **Embeddings**: Static embeddings might not distinguish the nuanced differences, but dynamic embeddings can capture the slight contextual shifts.

### Formal vs. Informal Lexical Choices

**Scenario**: Shifts between formal and informal language within a discourse.

**Example**: A speech includes both ASSISTANCE and HELP, or TERMINATE and END.

**Category**: **Register Variation**

**Characteristics**:
- **Formality Levels**: Variations in lexical choices can indicate shifts in formality or audience.
- **Implication**: Can reflect the context or the intended formality level of the audience.
- **Tool Insights**:
  - **Dictionaries**: Formality levels can be identified using resources like LIWC.
  - **Embeddings**: Contextual embeddings (BERT, ELMo) can capture shifts in formality based on surrounding words.

### Technical vs. Layman Terms

**Scenario**: Use of technical jargon vs. common language.

**Example**: A scientific paper vs. a popular science article. Using MYOCARDIAL INFARCTION in one and HEART ATTACK in the other.

**Category**: **Domain-Specific Language**

**Characteristics**:
- **Technical Jargon**: Indicates the level of expertise expected from the audience.
- **Implication**: Using technical terms can make a text more precise but harder to understand for a general audience.
- **Tool Insights**:
  - **WordNet and Specialized Dictionaries**: Identify domain-specific terms.
  - **Embeddings**: Dynamic embeddings can adjust for technical vs. common usage based on context.

### Emotional vs. Neutral Language

**Scenario**: Shifts between emotionally charged words and neutral terms.

**Example**: Describing an event with words like DEVASTATING vs. SIGNIFICANT.

**Category**: **Emotional Tone**

**Characteristics**:
- **Emotional Charge**: Words with strong emotional connotations vs. neutral descriptions.
- **Implication**: Can influence the reader's emotional response and engagement.
- **Tool Insights**:
  - **LIWC**: Measures emotional tone and psychological states.
  - **Dynamic Embeddings**: Capture the emotional context within sentences.

### Concrete vs. Abstract Language

**Scenario**: Use of concrete nouns vs. abstract concepts.

**Example**: APPLE vs. FREEDOM

**Category**: **Concrete vs. Abstract Language**

**Characteristics**:
- **Specificity**: Concrete words make the text more tangible, while abstract words can convey broader ideas.
- **Implication**: The choice can affect the clarity and imagery in the discourse.
- **Tool Insights**:
  - **WordNet**: Can identify the nature of words (concrete vs. abstract).
  - **Embeddings**: Contextual embeddings can show how abstract concepts are grounded in specific contexts.

### Frequent vs. Rare Words

**Scenario**: Use of common vs. rare words.

**Example**: HOUSE vs. MANSE.

**Category**: **Lexical Frequency**

**Characteristics**:
- **Accessibility**: Frequent words are more easily understood by a wider audience.
- **Implication**: Rare words can add sophistication but may reduce readability.
- **Tool Insights**:
  - **Corpora (COCA, BNC)**: Provide frequency information.
  - **Embeddings**: Frequency impacts the quality of static embeddings, while dynamic embeddings can better handle rare words in context.

### Specificity vs. Generality

**Scenario**: Use of specific terms vs. general terms.

**Example**: POODLE vs. DOG

**Category**: **Specificity**

**Characteristics**:
- **Detail Level**: Specific terms provide more detail, while general terms can be more inclusive.
- **Implication**: Specificity can clarify or narrow the focus, while generality can broaden it.
- **Tool Insights**:
  - **WordNet**: Identifies hypernyms and hyponyms.
  - **Embeddings**: Contextual embeddings can adjust specificity based on context.
